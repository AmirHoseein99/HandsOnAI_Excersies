```python
import tensorflow as ts
from tensorflow import keras
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

Q9 - Build your own CNN from scratch and try to achieve the highest possible accuracy on MNIST

```python
mnist_dataset = keras.datasets.mnist
(X_train_full, y_train_full), (X_test, y_test) = mnist_dataset.load_data()
```

```python
X_valid, X_train = X_train_full[:5000] / 255.0, X_train_full[5000:] / 255.0
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
X_valid = X_valid.reshape(5000, 28, 28, 1)
X_train = X_train.reshape(55000, 28, 28, 1)
```

```python
X_test = X_test.reshape(10000, 28, 28, 1)
```

```python
X_test.shape
```

```python
cnnModel = keras.models.Sequential([
      keras.layers.Conv2D(64, 7, activation="relu", padding="same",
                          input_shape=[28, 28, 1]),
      keras.layers.MaxPooling2D(2),
      keras.layers.Conv2D(128, 3, activation="relu", padding="same"),
      keras.layers.Conv2D(128, 3, activation="relu", padding="same"),
      keras.layers.MaxPooling2D(2),
      keras.layers.Conv2D(256, 3, activation="relu", padding="same"),
      keras.layers.Conv2D(256, 3, activation="relu", padding="same"),
            keras.layers.MaxPooling2D(2),
      keras.layers.Flatten(input_shape=[28, 28]),
      keras.layers.Dense(128, activation="relu"),
      keras.layers.Dropout(0.5),
      keras.layers.Dense(64, activation="relu"),
      keras.layers.Dropout(0.5),
      keras.layers.Dense(10, activation="softmax")
])
```

```python
cnnModel.summary()
```

```python
earlystop_cb = keras.callbacks.EarlyStopping(patience=10)
callbacks = [earlystop_cb]
```

```python
cnnModel.compile(
    loss="sparse_categorical_crossentropy",
    optimizer='sgd',
    metrics=['accuracy']
    )
```

```python
history = cnnModel.fit(X_train, y_train, epochs=100, validation_data=(X_valid, y_valid), callbacks=callbacks, verbose=1)
```

```python
cnnModel.evaluate(X_test, y_test)
```
